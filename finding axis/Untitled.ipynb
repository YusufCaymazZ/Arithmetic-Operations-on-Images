```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("kara.jpg")
```

```python
cv2.imshow("kara",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Matplotlib grafiini oluturun
plt.imshow(img_rgb, cmap='viridis')
plt.title('Grnt')
plt.axis('on')  # Eksenleri kapatmak istiyorsanz kullanabilirsiniz
plt.figure(figsize=(16, 16))
plt.show()
```

```python
# Kullancdan balang ve biti koordinatlarn al
start_x = int(input("Balang X koordinatn girin: "))
start_y = int(input("Balang Y koordinatn girin: "))
end_x = int(input("Biti X koordinatn girin: "))
end_y = int(input("Biti Y koordinatn girin: "))

# Belirtilen blgeyi se
selected_region = img[start_y:end_y, start_x:end_x]

# Seilen blgeyi gster
cv2.imshow("Seilen Blge", selected_region)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
