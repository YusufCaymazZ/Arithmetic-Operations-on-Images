```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("kara.jpg")
```

```python
cv2.imshow("kara",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Matplotlib grafiğini oluşturun
plt.imshow(img_rgb, cmap='viridis')
plt.title('Görüntü')
plt.axis('on')  # Eksenleri kapatmak istiyorsanız kullanabilirsiniz
plt.figure(figsize=(16, 16))
plt.show()
```

```python
# Kullanıcıdan başlangıç ve bitiş koordinatlarını al
start_x = int(input("Başlangıç X koordinatını girin: "))
start_y = int(input("Başlangıç Y koordinatını girin: "))
end_x = int(input("Bitiş X koordinatını girin: "))
end_y = int(input("Bitiş Y koordinatını girin: "))

# Belirtilen bölgeyi seç
selected_region = img[start_y:end_y, start_x:end_x]

# Seçilen bölgeyi göster
cv2.imshow("Seçilen Bölge", selected_region)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
